# Multiple Plots

We can have multiple plots in the same figure, and to achieve this we can, either use `Matplotlib` or `Seaborn`.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
```

## Using Matplotlib

```python
data = sns.load_dataset("iris")

plt.figure(figsize=(11, 3))
plt.subplot(121)
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

plt.subplot(122)
sns.lineplot(x="petal_length", y="petal_width", data=data)

plt.tight_layout()
```

## Using Seaborn

We have two methods in Seaborn to plot multiple axes on the same figure:

1. `FacetGrid()` can help us visualizing distribution of one variable, as well as the relationship between other variables.
2. `PairGrid()` can help us plotting pairwise relationships quickly.

## FacetGrid



```python
data = sns.load_dataset("iris")
grid = sns.FacetGrid(data, col="species")
grid.map(plt.plot, "sepal_width")
```

```python
data = sns.load_dataset("titanic")
grid = sns.FacetGrid(data, col="class", row="sex")
grid.map(sns.histplot, "alive")
grid.add_legend()
```

## PairGrid

```python
data = sns.load_dataset("iris")
vars = ["sepal_length", "sepal_width"] 

grid = sns.PairGrid(data, vars=vars, hue="species")
grid.map_diag(sns.histplot)
grid.map_offdiag(sns.scatterplot)
grid.add_legend()
```

```python
data = sns.load_dataset("iris")
x_vars = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
y_vars = ["species"]

grid = sns.PairGrid(data, x_vars=x_vars, y_vars=y_vars)
grid.map(sns.barplot)
```
