# Seaborn Basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Plotting a Simple Line



### Plotting with x-axis and y-axis

```python
x = np.array([1, 2, 3, 4])
y = x**3

# Seaborn has 3 parameters in plotting functions, x, y, and data.
# When we are plotting without data, we must specify the keyword arguments x and y.
# (⭕) plot(x=x, y=y)
# (❌) plot(x, y)

sns.lineplot(x=x, y=y)
```

### Plotting with `data` parameter

```python
df = pd.DataFrame(list(zip(x, y)), columns=["X", "y"])
print(df)

# When we are plotting with data (e.g., pandas.DataFrame)
# we could simply specify the column's name from the data to the keyword arguments x and y
#
# and Seaborn will set the xlabel and ylabel automatically

sns.lineplot(x="X", y="y", data=df)
```

## Using Seaborn with Matplotlib

```python
# loading dataset 
data = sns.load_dataset("iris")  # return pandas.DataFrame
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
  
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
```

## Seaborn Styles

There are five themes available in Seaborn, we can switch between them by using `set_style()`.

- darkgrid
- whitegrid
- dark
- white
- ticks

```python
sns.set_style("whitegrid")

data = sns.load_dataset("iris")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title('Title using Matplotlib Function')
```

```python
sns.set_style("dark")

data = sns.load_dataset("iris")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title('Title using Matplotlib Function')
```

## Changing the figure Size with `plt.figure()`

```python
width = 16
height = 4
plt.figure(figsize=(width, height))

data = sns.load_dataset("iris")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title('Title using Matplotlib Function')
```

# Scaling the plots

`sns.set_context()` can affect things like size of labels, size of lines, etc. The default context is `notebook`, and others are `paper`, `talk`, `poster`.

It is also possible to change the font scale by changing the `font_scale` argument.

```python
sns.set_context("paper", font_scale=2)
data = sns.load_dataset("iris")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title('Title using Matplotlib Function')
```

## Setting Styles Temporarily

```python
data = sns.load_dataset("iris")
sns.set_context("notebook")

def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

with sns.axes_style("whitegrid"):
    plt.subplot(211)
    plot()

plt.subplot(212)
plot()
```
