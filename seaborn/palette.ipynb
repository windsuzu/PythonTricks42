# Palette

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline
```

## Plotting Palette with `palplot` 

```python
palette = sns.color_palette()
sns.palplot(palette)
plt.title("Default Palette")
```

```python
palette = sns.color_palette("YlGn")
sns.palplot(palette)
plt.title("YlGn Colormap (Sequential)")

palette = sns.color_palette("coolwarm")
sns.palplot(palette)
plt.title("coolwarm Colormap (Diverging)")

palette = sns.color_palette("Pastel2")
sns.palplot(palette)
plt.title("Pastel2 Colormap (Qualitative)")
```

## Setting Palette

We can set the palette for any plot by simply calling `sns.set_palette()` or setting `palette` property in plotting methods.

```python
data = sns.load_dataset("iris")

sns.set_palette("Pastel1")
sns.lineplot(x="petal_length", y="petal_width", data=data)

sns.scatterplot(x="sepal_length", y="sepal_width", data=data, hue="sepal_width", palette="YlGn")
```

# Palette Category

We can categorize the colors according to 3 different situations:

1. **Qualitative**: Used when variable is categorical in nature, where each color needs to be **distinct**.
2. **Sequential**: Used when variable is numeric or ordered values, color moves **from lighter to darker**.
3. **Diverging**: Used for values are between the 2 extreme points. Consider a range of -10 to 10 so the value from -10 to 0 takes one color series and values from 0 to 10 take another.

## Sequential

![](..\assets\seaborn\sequential_colormaps.png)
![](..\assets\seaborn\sequential_colormaps_2.png)

```python
data = sns.load_dataset("iris")

# use `hue` to specify that the palette will affect on what feature.
sns.stripplot(x='species', y='sepal_width', hue="sepal_width", palette="Blues", data=data)
plt.legend([],[], frameon=False)
```
