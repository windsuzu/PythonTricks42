# Histograms

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn")

%matplotlib inline
```

```python
height_stats = np.random.default_rng(42).normal(160, 15, 1000)
height_stats[:10]
```

# Plotting Histograms

## Divide Data into Bins using Integers

```python
num_bin = 10
plt.hist(height_stats, bins=num_bin)
```

## Divide Data into Bins using Arrays

```python
interval_bin = [120, 130, 140, 150, 160, 170, 180, 190, 200]
plt.hist(height_stats, bins=interval_bin,
         edgecolor="black", lw=1)

plt.title("Height Stats")
plt.xlabel("Heights")
plt.ylabel("Counts")
plt.tight_layout()
```

# Add a Line of Mean

```python
mean = np.mean(height_stats)
mean
```

```python
interval_bin = [120, 130, 140, 150, 160, 170, 180, 190, 200]
plt.hist(height_stats, bins=interval_bin,
         edgecolor="black", lw=1)


plt.axvline(mean, c="orange", lw=5, label="Height Mean")
plt.legend()

plt.title("Height Stats")
plt.xlabel("Heights")
plt.ylabel("Counts")
plt.tight_layout()
```

# Add Density Curves

A more efficient way to plot histograms + Density Curve is using `seaborn.displot()`.

```python
import scipy.stats as ss
density = ss.kde.gaussian_kde(height_stats)
```

```python
interval_bin = [120, 130, 140, 150, 160, 170, 180, 190, 200]
plt.hist(height_stats, bins=interval_bin,
         edgecolor="black", lw=1, density=True)

index = np.arange(120, 200)
plt.plot(index, 
         density.evaluate(index), 
         color="pink",
         lw=3,
         ls="--",
         label="Probability Density")

plt.axvline(np.mean(height_stats), c="orange", lw=5, label="Height Mean")
plt.legend()

plt.title("Height Stats")
plt.xlabel("Heights")
plt.ylabel("Probability Density")
plt.tight_layout()
```
