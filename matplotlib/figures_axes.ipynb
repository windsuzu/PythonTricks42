# Multiple Figures and Axes

In Matplotlib, we have two approaches to create multiple `figures` and multiple `axes (plots)`.

- One is `object-oriented` interface
- Another is `Pyplot` interface

> âš¡ The definition of `axes` in `Matplotlib` is not the **plural form of axis**, but refers to `plots`.

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

## Pyplot interface

When we use `pyplot` to control `figures` and `axes`, we can create and switch between `figures` using:

``` py
plt.figure(1)  # create figure 1
plt.plot(...)

plt.figure(2)  # create and switch to figure 2
plt.figure(1)  # switch back to figure 1
```

For creating and switching between `axes`, we use `pyplot.subplot(num_of_rows, num_of_cols, current_position)`. You can also pass three parameters without comma separation, but it is less recommended.

``` py
plt.subplot(2, 2, 1)  # first create a 2x2 figure and work on the 1st block

plt.subplot(2, 2, 2)  # switch to the 2nd block in this 2x2 figure

plt.subplot(212)  # convert the figure to a 2x1 shape, and work on the 2nd block. 
                  # It is like (2, 2, [3, 4]) in Matlab
```

We can also assign `plt.subplot()` into an `axes` variable. Therefore, other `axes` can share the same x or y-axis using `sharex` or `sharey` parameters. 

``` py
ax = plt.subplot(2, 1, 1)
plt.plot([1, 2, 3, 4])
plt.axis = [-10, 10, -10, 10]

plt.subplot(2, 1, 2)
plt.plot([2, 3, 4, 5], sharex=ax, sharey=ax)
```

````python
ax = plt.subplot(2, 1, 1)
plt.plot([1, 2, 3, 4])
plt.axis([-10, 10, -10, 10])  # x-low, x-high, y-low, y-high
plt.title("Axes 1")


plt.subplot(2, 1, 2, sharex=ax, sharey=ax)
plt.plot([5, 7, 9, 11])
plt.title("Axes 2 (use the same x-y from Axes 1)")
plt.tight_layout(True)
````

### Multiple Axes (Plots)



````python
t1 = np.arange(0.0, 100.0, 0.1)
t2 = np.arange(10.0, 0.0, -0.1)
t3 = (np.arange(-10.0, 10.0, 0.1)) ** 2

ax1 = plt.subplot(2, 2, 1)
plt.plot(t1, marker="o", lw="5", c="orange")
plt.title("Axes 1")
plt.xticks([])

plt.subplot(2, 2, 2, sharey=ax1)
plt.plot(t2, marker="^", lw="3", c="pink")
plt.title("Axes 2")
plt.yticks([])
plt.xticks([])

plt.subplot(212)
plt.plot(t3)
plt.title("Axes 3")
````

````python
plt.figure(1)
plt.plot([1, 2, 3, 4], c="black", lw=10)
plt.title("Figure 1")

plt.figure(2)
plt.subplot(2, 1, 1)
plt.title("Figure 2")
plt.plot([2, 2, 3, 3], c="red", lw=10)
plt.subplot(2, 1, 2)
plt.plot([3, 3, 2, 2], c="red", lw=10)


plt.savefig("pic2.png")  # save figure 2 first

plt.figure(1)
plt.savefig("pic1.jpg")  # save figure 1 after switch to it
````

## Object-oriented interface

When we control `figures` and `axes` in the **object-oriented style**, we have to pre-define the needed `figures` and `axes`.

``` py
fig1, ax = plt.subplots()  # create 1 new single figure and 1 new axes on it


fig2, axs = plt.subplots(2, 1)  # create another 1 figure but new 2 axes on it
fig2, (ax1, ax2) = plt.subplots(2, 1)  # same


fig3, axs = plt.subplots(2, 2)  # 1 figure and 4 axes
# axs = [[plot, plot], [plot, plot]]
```

### Multiple Axes

````python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(np.sin(np.arange(0, 20)), c="red")
axs[0, 1].plot(np.cos(np.arange(0, 20)), c="orange")
axs[1, 0].plot(np.tan(np.arange(0, 20)), c="blue")
axs[1, 1].plot(np.log(np.arange(0, 20)), c="green")
````

### Multiple Figures

````python
fig1, ax = plt.subplots()
ax.plot([1, 2, 3, 4], "o--r")

fig2, axs = plt.subplots(2, 1)
axs[0].plot([1, 2, 3, 4], "o--b")
axs[0].set_title("axes 1")
axs[1].plot([4, 3, 2, 1], "o--g")
axs[1].set_title("axes 2")

fig2.set_tight_layout(True)

fig1.savefig("fig 1.jpg")
fig2.savefig("fig 2.jpg")
````
