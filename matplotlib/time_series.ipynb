# Plot Time Series

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dates = np.arange(np.datetime64("2021-01-01"), np.datetime64("2021-01-22"))
dates[:10]
```

```python
prices = np.random.default_rng(42).normal(500, 30, len(dates))
prices[:10]
```

## Plotting

```python
plt.plot_date(dates, prices, 
              ls="solid",
              c="orange",
              marker="^",
              markersize=10)
```

## Date Format

```python
import matplotlib.dates as mdates

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%a, %d %m"))
plt.plot_date(dates, prices, ls="solid")

plt.grid()
plt.tight_layout()
```

## Major Tick on X-axis

```python
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%a, %d %m"))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.plot_date(dates, prices, ls="solid")

plt.grid()
plt.tight_layout()
```

## Minor Tick on X-axis

```python
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%a, %d %m"))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.gca().xaxis.set_minor_locator(mdates.DayLocator())
plt.plot_date(dates, prices, ls="solid")

plt.grid()
plt.tight_layout()
```
