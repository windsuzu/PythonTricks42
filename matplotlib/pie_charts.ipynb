# Pie Charts

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Plotting Pie Charts

```python
partition = [40, 60]
labels = ["Bond", "Stock"]
colors = ["red", "green"]

plt.pie(partition, 
        labels=labels, 
        colors=colors,
        shadow=True,
        wedgeprops={'edgecolor': 'black', 'lw': 3},
        textprops={"fontsize": 16}, 
        autopct="%1.1f%%",
        pctdistance=0.5
        )

plt.title("My Portfolio")
plt.tight_layout()
```

## Explode and StartAngle



```python
plt.style.use("seaborn-pastel")
grade = ["A", "B", "C", "D", "E"]
number = [10, 18, 23, 8, 5]
explode = [0.1, 0, 0, 0, 0]

plt.pie(number, 
        labels=grade,
        shadow=True,
        autopct="%1.1f%%",
        pctdistance=0.6,
        startangle=90,
        explode=explode
        )

plt.title("Test Grade")
plt.tight_layout()
```
